### Just a test make sure the Monte Carlo integration is working.

```python
W1answer = 2.0294058703700369575
con = []
npts_list = [2**n for n in range(15)]
cutoff = 100
test = []
for npts in npts_list:
    test.append(1/np.sqrt(npts))
    con.append(abs(W1answer - monte_carlo(W1, cell_vecs, npts, cutoff)))
plt.loglog(npts_list, con, npts_list, test)
plt.show()
```
