```python
import numpy as np
from BZI.plots import PlotBandStructure, PlotVaspBandStructure
from BZI.pseudopots import *
```

```python
centering_type = "face"
lat_consts = [Al_lat_const]*3
lat_angles = [np.pi/2]*3
lattice = Lattice(centering_type, lat_consts, lat_angles)
Al_pff = [0.0179, 0.0562]
Al_cutoff = 4*(2*np.pi/Al_lat_const)**2
Al_atomic_basis = [[0.,0.,0.]]

Al_P = EmpiricalPP(lattice, Al_pff, Al_cutoff, Al_atomic_basis)
```

```python
Al_params = {"materials_list": ["Al"],
             "PPlist": [Al_P],
             "PPargs_list": [{}],
             "lattice": lattice,
             "npts": 20,
             "neigvals": 10,
             "energy_shift": 0.0,
             "energy_limits": [-5,30],
             "show": True
}
PlotBandStructure(**Al_params)
```
